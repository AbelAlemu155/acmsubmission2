```python
#  preprocess the data 
import os
import pandas as pd
ama_principles = {'1': 'Principle I - Competence and Compassion'
        , '2': "Principle II - Professional Integrity", 
        '3': "Principle III - Legal Compliance", 
        "4": "Principle IV - Patient Privacy", 
        "5": "Principle V - Continued Education", 
        "6": "Principle VI - Professional Choice", 
        "7": "Principle VII - Community Health", 
        "8": "Principle VIII - Patient Priority", 
        "9": "Principle IX - Universal Access"
}

med_safety_path= "./med-safety-bench/datasets/test/gpt4"
trident_path_medical= "./Trident/med_final.jsonl"
trident_path_law= "./Trident/law_final.jsonl"
def process_med_safety():
   
    files=  os.listdir(med_safety_path)

    # print(len(files))
    # print(files)

    med_safety_df= pd.DataFrame()
    for f_name in files:
        full_path= f"{med_safety_path}/{f_name}"
        ethical_refs= ama_principles[f_name[-5]]
        df = pd.read_csv(full_path, usecols=["harmful_medical_request", "safe_response"])
        df['ethical_refs']=ethical_refs
        df['domain']= 'med'
        med_safety_df= pd.concat([med_safety_df,df], ignore_index=True)

    med_safety_df = med_safety_df.rename(columns={"harmful_medical_request": "harmful_prompt", "safe_response":"safe_reply"})
    return med_safety_df 
```

```python
import csv
import os
import re
import time
from openai import OpenAI
from openai import RateLimitError

def generate_prompts_safe(df, client, few_shot_examples_law, few_shot_examples_med,csv_file="rewritten_prompts_safe.csv"):
    file_exists = os.path.isfile(csv_file)

    # open CSV in append mode once (better performance)
    with open(csv_file, mode="a", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["rewritten_prompt", "ethical_refs", "domain", "harmful_prompt"])

        if not file_exists:
            writer.writeheader()

        for idx, row in df.iterrows():
            harmful_prompt = row["harmful_prompt"]
            rule_violated = row["ethical_refs"]
            domain = row["domain"]
            few_s= few_shot_examples_med if domain=='med' else few_shot_examples_law
            try:
                completion = client.chat.completions.create(
                    model="moonshotai/Kimi-K2-Instruct:novita",
                    messages=[
                        {
                            "role": "user",
                            "content": get_rewritten_prompt_safe(harmful_prompt, rule_violated,few_s)
                        }
                    ],
                )

                generated_output = completion.choices[0].message.content
                prompts = re.findall(r"\[\[(.*?)\]\]", generated_output, re.DOTALL)

                if len(prompts) != 5:
                    print(f"Row {idx}: expected 5 prompts but found {len(prompts)}")

                for p in prompts:
                    writer.writerow({
                        "rewritten_prompt": p.strip(),
                        "ethical_refs": rule_violated,
                        "domain": domain,
                        "harmful_prompt": harmful_prompt
                    })

            except RateLimitError:
                print("Rate limit hit — sleeping 10 seconds")
                time.sleep(10)
                continue  # retry next iteration (or add retry logic)

            except Exception as e:
                print(f"Error on row {idx}: {e}")
                # optionally continue or break
                continue

    print("CSV writing complete")
```

```python
import os 
def sample_anchors(overrefusal_df, benign_df, harmful_anchor_out='h_anchor.csv', safe_anchor_out="s_anchor.csv", over_ref_anchor_out="o_anchor.csv"):
    if os.path.exists(harmful_anchor_out):
        return pd.read_csv(over_ref_anchor_out), pd.read_csv(safe_anchor_out), pd.read_csv(harmful_anchor_out)
    overrefusal_df_interim= overrefusal_df.copy()
    overrefusal_df_interim= overrefusal_df_interim.drop_duplicates(subset=['rewritten_prompt'], keep='first')
    overrefusal_df_interim= overrefusal_df_interim[overrefusal_df_interim.rewritten_prompt.str.len() >= 20]
    print(overrefusal_df_interim.head())
    benign_df_interim= benign_df.copy()
    benign_df_interim= benign_df_interim.drop_duplicates(subset=['rewritten_prompt'], keep='first')
    benign_df_interim= benign_df_interim[benign_df_interim.rewritten_prompt.str.len() >= 20]
    harmful_df= overrefusal_df_interim[['harmful_prompt', 'domain', 'ethical_refs']].drop_duplicates(subset=['harmful_prompt'], keep='first')
    overrefusal_anchors= sampler(overrefusal_df_interim)
    benign_anchors= sampler(benign_df_interim)
    harmful_anchors= sampler(harmful_df)
    overrefusal_anchors.to_csv(over_ref_anchor_out)
    benign_anchors.to_csv(safe_anchor_out)
    harmful_anchors.to_csv(harmful_anchor_out)
    return overrefusal_anchors, benign_anchors, harmful_anchors
def sampler(df):
    samples = []

    # get unique ethical refs
    for ref in df['ethical_refs'].unique():
        
        sub = df[df['ethical_refs'] == ref]

        # sample med (up to 10)
        med_subset = sub[sub['domain'] == 'med']
        med_sample = med_subset.sample(
            n=min(10, len(med_subset)),
            random_state=42
        )

        # sample law (up to 2)
        law_subset = sub[sub['domain'] == 'law']
        law_sample = law_subset.sample(
            n=min(2, len(law_subset)),
            random_state=42
        )

        samples.append(med_sample)
        samples.append(law_sample)

    # combine all samples
    result = pd.concat(samples, ignore_index=True)

    print(result.head())
    return result
    
        
```

```python
def get_hidden_states(model, tokenizer, full_prompt_list):
    hidden_state_list = []

    with torch.no_grad():
        progress_bar = tqdm(full_prompt_list, desc="Calculating hidden states")
        for full_prompt in progress_bar:
            inputs = tokenizer(full_prompt, return_tensors="pt", padding=True).to(
                model.device
            )
            outputs = model(**inputs, output_hidden_states=True)
            # Get the last hidden state of the last token for each sequence
            # We use -1 to index the last layer, and -1 again to index the hidden state of the last token
            hidden_state_list.append(outputs.hidden_states[-1][:, -1, :])  # [1, D]
    hidden_state_list = torch.cat(hidden_state_list, dim=0)  # [N, D]
    return hidden_state_list


def get_hidden_states_batch(model, tokenizer, full_prompt_list, batch_size=8):
    hidden_state_list = []
    num_samples = len(full_prompt_list)

    with torch.no_grad():
        for i in range(0, num_samples, batch_size):
            batch = full_prompt_list[i:i+batch_size]
            inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(model.device)
            outputs = model(**inputs, output_hidden_states=True)
            # Get the last hidden state of the last token for each sequence in the batch
            hidden_state_list.append(outputs.hidden_states[-1][:, -1, :])  # [batch_size, D]

    hidden_state_list = torch.cat(hidden_state_list, dim=0)  # [N, D]
    return hidden_state_list
```

```python
def replace_nan(x):
    if torch.isnan(x).any():
        non_nan_mask = ~torch.isnan(x)  
        non_nan_values = x[non_nan_mask]  
        non_nan_mean = non_nan_values.mean()  
        x[torch.isnan(x)] = non_nan_mean
    return x
```

```python
import json

import pandas as pd
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

def load_model_and_tokenizer(model_name_or_path, use_flash_attn=True):
    
    model = AutoModelForCausalLM.from_pretrained(
        model_name_or_path,
        device_map="auto",
        low_cpu_mem_usage=True,
        torch_dtype=torch.float16,
        trust_remote_code=True,
       
    )
    print("Using FP16 and normal attention implementation...")

    tokenizer = load_tokenizer(model_name_or_path)
    return model, tokenizer


def load_tokenizer(model_name_or_path):
    tokenizer = AutoTokenizer.from_pretrained(model_name_or_path, trust_remote_code=True)
    if tokenizer.pad_token is None:
        if tokenizer.eos_token is not None:
            tokenizer.pad_token = tokenizer.eos_token
            tokenizer.pad_token_id = tokenizer.eos_token_id
        else:
            tokenizer.pad_token = '<|endoftext|>'
            tokenizer.pad_token_id = tokenizer.eod_id
    if tokenizer.clean_up_tokenization_spaces:
        print(
            "WARNING: tokenizer.clean_up_tokenization_spaces is by default set to True. "
            "This will break the attack when validating re-tokenization invariance. Setting it to False..."
        )
        tokenizer.clean_up_tokenization_spaces = False

    # If the chat template is not available, manually set one
    # Some older models do not come with a chat template in their tokenizer
    # Alternatively, you can also modify `tokenizer_config.json` file in the model directory, if you locally have access to it
    if not tokenizer.chat_template:
        print(
            f"The tokenizer of {model_name_or_path} does not come with a chat template. Dynamically setting one..."
        )
        if "harmbench-llama2-13b-cls" in model_name_or_path:
            # If you are sure that the model does not require a chat template, you can skip this step like this
            print(
                "HarmBench-Llama-2-13b-cls does not require a chat template. Skipped."
            )
        elif "vicuna-7b-v1.5" in model_name_or_path:
            # Otherwise, please implement the chat template manually in Jinja language like this
            # No indentation and newlines are allowed. Please make it a single line
            tokenizer.chat_template = "{% if messages[0]['role'] == 'system' %}{% set loop_messages = messages[1:] %}{% set system_message = messages[0]['content'].strip() + '' %}{% else %}{% set loop_messages = messages %}{% set system_message = '' %}{% endif %}{{ bos_token + system_message }}{% for message in loop_messages %}{% if (message['role'] == 'user') != (loop.index0 % 2 == 0) %}{{ raise_exception('Conversation roles must alternate user/assistant/user/assistant/...') }}{% endif %}{% if message['role'] == 'user' %}{{ 'USER: ' + message['content'].strip() + '' }}{% elif message['role'] == 'assistant' %}{{ 'ASSISTANT: ' + message['content'].strip() + eos_token + '' }}{% endif %}{% if loop.last and message['role'] == 'user' and add_generation_prompt %}{{ 'ASSISTANT:' }}{% endif %}{% endfor %}"
        elif "qwen" in model_name_or_path:
            tokenizer.chat_template = "{% for message in messages %}{% if loop.first and messages[0]['role'] != 'system' %}{{ '<|im_start|>system\nYou are a helpful assistant<|im_end|>\n' }}{% endif %}{{'<|im_start|>' + message['role'] + '\n' + message['content'] + '<|im_end|>' + '\n'}}{% endfor %}{% if add_generation_prompt %}{{ '<|im_start|>assistant\n' }}{% endif %}"
        elif "baichuan" in model_name_or_path:
            tokenizer.chat_template = "{% for message in messages %}{% if message['role'] == 'user' %}{{ '<reserved_106>' + message['content'] }}{% elif message['role'] == 'assistant' %}{{ '<reserved_107>' + message['content'] + '</s>'}}{% endif %}{% endfor %}{% if add_generation_prompt and messages[-1]['role'] != 'assistant' %}{{ '<reserved_107>' }}{% endif %}"
        elif "llama2-70b" in model_name_or_path:
            tokenizer.chat_template = "{% if messages[0]['role'] == 'system' %}{% set system_message = '<<SYS>>\n' + messages[0]['content'] | trim + '\n<</SYS>>\n\n' %}{% set messages = messages[1:] %}{% else %}{% set system_message = '' %}{% endif %}{% for message in messages %}{% if (message['role'] == 'user') != (loop.index0 % 2 == 0) %}{{ raise_exception('Conversation roles must alternate user/assistant/user/assistant/...') }}{% endif %}{% if loop.index0 == 0 %}{% set content = system_message + message['content'] %}{% else %}{% set content = message['content'] %}{% endif %}{% if message['role'] == 'user' %}{{ bos_token + '[INST] ' + content | trim + ' [/INST]' }}{% elif message['role'] == 'assistant' %}{{ ' ' + content | trim + ' ' + eos_token }}{% endif %}{% endfor %}"
        else:
            raise ValueError(
                f"The chat template for the tokenizer of {model_name_or_path} is not available. "
                "To avoid unexpected behavior, it cannot proceed with the default chat template. "
                "Please implement it manually in `load_model_and_tokenizer()`, `utils.py`."
            )
    return tokenizer
```

```python
def apply_chat_template(tokenizer, prompts, system_prompt=None, add_generation_prompt=False):
    full_prompt_list = []
    for _, prompt in enumerate(prompts):
        if prompt is None:
            prompt = ""
        prompt = str(prompt) 
        if system_prompt is not None:
            messages = [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": prompt},
            ]
        else:
            messages = [{"role": "user", "content": prompt}]
        full_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=add_generation_prompt)
        full_prompt_list.append(full_prompt)

    return full_prompt_list
```

```python
from sklearn.decomposition import PCA
def pca_reduce_dimensions(hidden_states, n_components):
    pca = PCA(n_components=n_components)
    pca.fit(hidden_states)
    return pca
```

```python
from copy import deepcopy
from tqdm import tqdm
import numpy as np
def compute_score(model_name,model_path, overrefusal_df,overrefusal_anchors, benign_anchors, harmful_anchors,output_dir="hidden_states",seed=42 , batch_size= 4):
    model, tokenizer = load_model_and_tokenizer(model_path)
    overrefusal_df_f= overrefusal_df[~overrefusal_df["rewritten_prompt"].isin(overrefusal_anchors["rewritten_prompt"])]
    model.eval()
    df_anchors = {}

    df_anchors['benign'] = benign_anchors.rename(columns={'rewritten_prompt': 'prompt'})
    df_anchors['harmful'] = harmful_anchors.rename(columns={'harmful_prompt':'prompt'})
    df_anchors['overrefusal'] = overrefusal_anchors[['rewritten_prompt', 'domain', 'ethical_refs']].rename(columns={'rewritten_prompt': 'prompt'})

    try:
        
        hidden_states_anchors = {}
        for anchor, anchor_data in df_anchors.items():
            anchor_hidden_path = os.path.join(output_dir, f'{anchor}.pt')
            if os.path.exists(anchor_hidden_path):
                print(f'loading {anchor} hidden states from {anchor_hidden_path}')
                hidden_states = torch.load(anchor_hidden_path).cpu()
            else:
                full_prompts = apply_chat_template(tokenizer, anchor_data.prompt)
                hidden_states = get_hidden_states(model, tokenizer, full_prompts).cpu()
            hidden_states = replace_nan(hidden_states)
            hidden_states_anchors[anchor] = hidden_states
        
        hidden_states_anchor_cat = torch.cat(
                [hidden_states_anchors['benign'], hidden_states_anchors['harmful']], dim=0
            )
        pca = pca_reduce_dimensions(hidden_states_anchor_cat.cpu().numpy(), 2)
    except Exception as e:
        print(f"==>> {model_name} for generating vectors failed with error: {e}")
        return
    
    reduced_refusal = pca.transform(hidden_states_anchors['overrefusal'].cpu().numpy())
    refusal_point = np.mean(reduced_refusal, axis=0)
    # Also compute harmful point: center of harmful anchor data
    reduced_harmful = pca.transform(hidden_states_anchors['harmful'].cpu().numpy())
    harmful_point = np.mean(reduced_harmful, axis=0)

    refusal_point = torch.tensor(refusal_point).to(model.device)
    harmful_point = torch.tensor(harmful_point).to(model.device)
    print(f"==>> refusal_point_tensor: {refusal_point}")
    print(f"==>> harmful_point_tensor: {harmful_point}")

    over_refusal_direction = refusal_point - harmful_point
    over_refusal_direction = over_refusal_direction / torch.norm(over_refusal_direction)
    print(f"==>> over_refusal_direction: {over_refusal_direction}")

    mean = torch.tensor(pca.mean_).to(model.device)
    components = torch.tensor(pca.components_).to(model.device)
    overrefusal_df_f = overrefusal_df_f.rename(columns={'rewritten_prompt': 'prompt', 'harmful_prompt': 'toxic_seed'})
    for category in overrefusal_df_f['ethical_refs'].unique():
        cate_data= overrefusal_df_f[overrefusal_df_f['ethical_refs']==category]
        score_list = []
        hidden_or_list = []
        hidden_toxic_list = []
        num_samples = len(cate_data)
        full_prompts_or = apply_chat_template(tokenizer, cate_data.prompt)
        full_prompts_toxic = apply_chat_template(tokenizer, cate_data.toxic_seed)
        with torch.no_grad():
            for idx in tqdm(range(0, num_samples, batch_size)):
                if idx + batch_size > num_samples:
                    step_size = num_samples - idx
                else:
                    step_size = batch_size

                batch_or = full_prompts_or[idx:idx+step_size]
                inputs_or = tokenizer(batch_or, return_tensors="pt", padding=True, truncation=True).to(model.device)
                outputs_or = model(**inputs_or, output_hidden_states=True)
                last_token_indices_or = inputs_or.attention_mask.sum(dim=1) - 1  # 计算有效token位置
                hidden_or = outputs_or.hidden_states[-1][
                    torch.arange(step_size), 
                    last_token_indices_or    
                ] 
                hidden_or = replace_nan(hidden_or)
                hidden_or_list.append(hidden_or.cpu())

                batch_toxic = full_prompts_toxic[idx:idx+step_size]
                inputs_toxic = tokenizer(batch_toxic, return_tensors="pt", padding=True, truncation=True).to(model.device)
                outputs_toxic = model(**inputs_toxic, output_hidden_states=True)
               
                last_token_indices_toxic = inputs_toxic.attention_mask.sum(dim=1) - 1
                hidden_toxic = outputs_toxic.hidden_states[-1][
                    torch.arange(step_size),
                    last_token_indices_toxic
                ]  # [batch_size, D]
                hidden_toxic = replace_nan(hidden_toxic)
                hidden_toxic_list.append(hidden_toxic.cpu())
                vector_from_start_to_here = hidden_or - hidden_toxic
                vector_transformed = (vector_from_start_to_here - mean) @ components.T  # [N, 2]
                vector_transformed = vector_transformed / torch.norm(vector_transformed, dim=1, keepdim=True)
                score = torch.sum(vector_transformed * over_refusal_direction, dim=1)
                score_list.append(score.cpu())
        score_list = torch.cat(score_list, dim=0).cpu().numpy()

        save_data = deepcopy(cate_data)
        save_data['score'] = score_list
        save_data.to_csv(f'categories/{category}.csv', index=False)
```
